```python
# let's find out about outliers using plots
df3 = df1_dropRows
from matplotlib import pyplot as plt
import seaborn as sns
sns.boxplot(df3['PRICE'])
```

```python
df3.describe()
```

```python
# now let's try to remove outliers in Prices
mean_price = df3['PRICE'].mean()
std_price = df3['PRICE'].std()
top_range = mean_price + std_price*1.96
bot_range = mean_price - std_price*1.96

df4 = df3.copy()
df4 = df4.drop(df4[df4['PRICE'] > top_range].index)
df4 = df4.drop(df4[df4['PRICE'] < bot_range].index)
df4.head()
```
